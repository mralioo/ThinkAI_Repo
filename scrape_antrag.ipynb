```python
def save_as_markdown(content_list, file_name):
    """
    Save the extracted content as a Markdown file.

    :param content_list: A list of markdown content strings to save.
    :param file_name: The name of the file to save the content to.
    """
    with open(file_name, 'w', encoding='utf-8') as f:
        f.write("\n".join(content_list))
```

```python
def load_markdown(file_name):
    """
    Load the content from a Markdown file.

    :param file_name: The name of the file to load the content from.
    :return: The content of the file as a string.
    """
    with open(file_name, 'r', encoding='utf-8') as f:
        return f.read()
```

```python
def parse_text(text):
    # Dictionary to store the final result
    result = {}

    # Split the text into lines
    lines = text.split('\n')

    # Regex patterns
    topic_pattern = re.compile(r"^([^\*]+)\s*={2,}$")
    bullet_pattern = re.compile(r"^\*\s+\[([^\]]+)\]\(([^)]+)\s*\"([^\"]+)\"\)")

    current_topic = None

    for line in lines:
        line = line.strip()  # Trim whitespace
        topic_match = topic_pattern.match(line)
        bullet_match = bullet_pattern.match(line)

        if topic_match:
            current_topic = topic_match.group(1).strip()
            result[current_topic] = []
        elif bullet_match and current_topic:
            title = bullet_match.group(1).strip()
            link = bullet_match.group(2).strip()
            description = bullet_match.group(3).strip()
            result[current_topic].append({'title': title, 'link': link, 'description': description})

    return result
```
